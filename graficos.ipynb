```python
# Multiline string containing the data
data = '''
	POROUS	Xmethanog.
9/12/2018	3.69	50.4317111459969
11/13/2018	3.24	58.0062794348509
12/18/2018	2.67	47.2527472527473
2/22/2019	3.09	41.7974882260597
4/25/2019	2.08	49.25431711146
6/19/2019	2.78	40.6397174254317
'''


a = [] # Ex: Sample collect date
b = [] # Ex: Fractured
c = [] # Ex: Porous


# Split the data into lines
lines = data.strip().split('\n')

# Iterate over the lines, skipping the header
for line in lines[1:]:  # Skip the header line
    # Split each line by tab or whitespace
    parts = line.split()
    # Append the fractured and porous values to their respective lists
    a.append(parts[0])
    b.append(float(parts[1]))
    c.append(float(parts[2]))


print(f"a0= {a}\nb= {b}\nc= {c}")
```
